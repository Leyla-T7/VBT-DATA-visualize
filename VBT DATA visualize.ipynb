### Veri Görselleştirme

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
plt.figure()
plt.show()
```

```python
#BAR CHART
```

```python
df.groupby("user_city").agg({"total_amount":"mean"})
```

```python
plt.bar(data=df.groupby("user_city").agg({"total_amount":"mean"}).reset_index(),
       x="user_city",
       height="total_amount")
plt.title("table",size=12 , color="red")
plt.show()
```

```python
#SEABORN
```

```python
sns.set_style("whitegrid")#tabloya ızgaralar cızer
sns.barplot(data=df.groupby("user_city").agg({"total_amount":"mean"}).reset_index(),
            x="user_city",
            y="total_amount")
```

```python
#PİE CHART
```

```python
plt.pie(data=df.groupby("user_city").agg({"user_id":"count"}).reset_index(),
            labels="user_city",
            x="user_id")
plt.show()
```

```python
sns.countplot(data=df,
              x="user_city",
             color="pink")
```

```python
#SCATTER PLOT
```

```python
plt.scatter(data=df,
           x="user_age",
           y="total_amount",
           color="pink")

plt.xlabel("kullanıcı yası")
plt.ylabel("kredi kartı tuketım")
plt.show()
```

```python
#LİNE CHART
```

```python
user_df=df.groupby("card_delivery_date").agg({"user_id":"count"}).reset_index()
user_df.head()
```

```python
user_df["card_delivery_date"]=pd.to_datetime(user_df["card_delivery_date"])
```

```python
sns.lineplot(data=user_df,
            x="card_delivery_date",
            y="user_id")
```
